```python
import os
import shutil

# 대상 디렉터리 경로
target_directory = "/home/bxai1/BXAI/Project/NeuroTalk-decoder/Patients/voice_cut/"
destination_directory = "/home/bxai1/BXAI/Project/NeuroTalk-decoder/Patients/preprocessed/"

# 디렉터리 내 모든 폴더 이름 가져오기
for folder_name in sorted(os.listdir(target_directory)):
    sub_path = os.path.join(target_directory, folder_name)  # Voice cut의 subject 폴더
    patient_list = sorted(os.listdir(destination_directory))  # Preprocessed의 Patient 리스트 정렬

    # patient_number 초기화
    selected_patient_number = None

    # 대상 patient_number 찾기
    for patient_number in patient_list:
        if patient_number.find(folder_name[-3:]) != -1:
            selected_patient_number = patient_number
            break
    else:
        # 겹치는 이름이 없는 경우 전체 for문의 다음 단계로 넘어감
        print(f"'{folder_name}'에 해당하는 환자 폴더를 찾지 못했습니다. 다음 단계로 넘어갑니다.")
        continue

    # 지정된 patient_number로 작업 수행
    patient_path = os.path.join(destination_directory, selected_patient_number)  # 정렬한 리스트의 경로
    about_voice_path = os.path.join(patient_path, "about_voice")  # Preprocessed에서 about voice 경로 생성

    if not os.path.isdir(about_voice_path):
        os.makedirs(about_voice_path)  # about_voice가 존재하지 않는다면 생성
    else:
        print(f"{about_voice_path}가 이미 존재합니다")

    # 폴더인지 확인
    if os.path.isdir(about_voice_path):
        preprocessd_about_voice_subject_path = os.path.join(about_voice_path, folder_name)
        shutil.move(sub_path, preprocessd_about_voice_subject_path)
        new_name = folder_name + "_voice"
        new_folder_path = os.path.join(about_voice_path, new_name)

        # 폴더 이름 변경
        os.rename(preprocessd_about_voice_subject_path, new_folder_path)
        print(f"'{folder_name}' -> '{new_name}'로 변경됨.")
```
