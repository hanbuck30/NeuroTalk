```python
import os
import torchaudio
import pandas as pd
import numpy as np

def wav_to_csv(wav_path, csv_path):
    # WAV 파일 읽기
    waveform, sample_rate = torchaudio.load(wav_path)

    # 스테레오 -> 모노 변환 (두 채널의 평균)
    if waveform.size(0) > 1:  # 만약 스테레오라면
        waveform = waveform.mean(dim=0, keepdim=True)

    # 샘플을 1차원 numpy 배열로 변환
    waveform_np = waveform.numpy().flatten()

    # DataFrame 생성
    df = pd.DataFrame(waveform_np, columns=['Sample'])

    # CSV로 저장
    df.to_csv(csv_path, index=False)
```

```python
import csv
import wave
import numpy as np

def csv_to_wav(csv_file, wav_file, mic_number, mic_sr):
    """
    Converts a CSV file containing audio data to a WAV file.
    
    Args:
        csv_file (str): The path to the input CSV file.
        wav_file (str): The path to the output WAV file.
        mic_number (int): The microphone mode, 1 for mono and 2 for stereo.
        mic_sr (int): The sampling rate of the microphone.
    
    Raises:
        ValueError: If mic_number is not 1 or 2.
    """
    # Read audio data from the CSV file
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the first row
        data = [float(row[0]) for row in reader]

    # Create a WAV file
    with wave.open(wav_file, 'wb') as wav:
        wav.setnchannels(1)  # Mono channel
        wav.setsampwidth(2)  # 16-bit sample width
        if mic_number == 1:
            wav.setframerate(mic_sr)  # Set sample rate for mic 1
        elif mic_number == 2:
            wav.setframerate(mic_sr * 2)  # Set sample rate for mic 2
        else:
            raise ValueError("Invalid mic_number. Please enter 1 or 2.")
        
        # Convert data to 16-bit integers and write to the WAV file
        audio_data = np.array(data).astype(np.int16)
        wav.writeframes(audio_data.tobytes())
```
